```python
import os
import xml.etree.ElementTree as ET

image_dir = '/content/data/images'
annotation_dir = '/content/data/annotations'
output_label_dir = '/content/data/labels'

os.makedirs(output_label_dir, exist_ok=True)

# Mapping for face mask dataset
label_map = {
    'with_mask': 0,
    'mask_weared_incorrect': 1,
    'without_mask': 2
}

for xml_file in os.listdir(annotation_dir):
    if not xml_file.endswith(".xml"):
        continue

    tree = ET.parse(os.path.join(annotation_dir, xml_file))
    root = tree.getroot()

    img_filename = root.find('filename').text
    img_path = os.path.join(image_dir, img_filename)

    # Get image size
    size = root.find('size')
    w = int(size.find('width').text)
    h = int(size.find('height').text)

    yolo_lines = []

    for obj in root.findall('object'):
        class_name = obj.find('name').text
        if class_name not in label_map:
            continue
        class_id = label_map[class_name]

        bbox = obj.find('bndbox')
        xmin = int(bbox.find('xmin').text)
        ymin = int(bbox.find('ymin').text)
        xmax = int(bbox.find('xmax').text)
        ymax = int(bbox.find('ymax').text)

        # YOLO format
        x_center = ((xmin + xmax) / 2) / w
        y_center = ((ymin + ymax) / 2) / h
        box_width = (xmax - xmin) / w
        box_height = (ymax - ymin) / h

        yolo_lines.append(f"{class_id} {x_center:.6f} {y_center:.6f} {box_width:.6f} {box_height:.6f}")

    # Write label file
    txt_filename = xml_file.replace('.xml', '.txt')
    with open(os.path.join(output_label_dir, txt_filename), 'w') as f:
        f.write('\n'.join(yolo_lines))

```

```python
import os
import shutil
import random

# Source paths
images_path = '/content/data/images'
labels_path = '/content/data/labels'

# Output paths
train_img_dir = '/content/data/images/train'
val_img_dir = '/content/data/images/val'
train_lbl_dir = '/content/data/labels/train'
val_lbl_dir = '/content/data/labels/val'

# Create directories
os.makedirs(train_img_dir, exist_ok=True)
os.makedirs(val_img_dir, exist_ok=True)
os.makedirs(train_lbl_dir, exist_ok=True)
os.makedirs(val_lbl_dir, exist_ok=True)

# Split ratio
split_ratio = 0.8  # 80% train, 20% val

image_files = [f for f in os.listdir(images_path) if f.endswith(('.jpg', '.jpeg', '.png'))]
random.shuffle(image_files)

split_index = int(len(image_files) * split_ratio)
train_files = image_files[:split_index]
val_files = image_files[split_index:]

# Move images and corresponding labels
def move_files(file_list, img_dest, lbl_dest):
    for img_file in file_list:
        base_name = os.path.splitext(img_file)[0]
        label_file = base_name + '.txt'

        shutil.copy(os.path.join(images_path, img_file), os.path.join(img_dest, img_file))
        if os.path.exists(os.path.join(labels_path, label_file)):
            shutil.copy(os.path.join(labels_path, label_file), os.path.join(lbl_dest, label_file))

move_files(train_files, train_img_dir, train_lbl_dir)
move_files(val_files, val_img_dir, val_lbl_dir)
```

```python
data_yaml = """\
train: /content/data/images/train
val: /content/data/images/val

nc: 3
names: ['with_mask', 'mask_weared_incorrect', 'without_mask']
"""

with open('/content/data.yaml', 'w') as f:
    f.write(data_yaml)
```
